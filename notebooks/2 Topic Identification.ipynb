```python
fluid_topic = topic_results.get_topic_by_id(6)
phrases = wosis.find_phrases(fluid_topic)

# Example phrase identification
# Publications in this topic are related to fluid mechanistics
phrases.all_phrases.head()
```

```python
# Based on above, I have assigned the topic groups these overarching names
topic_names = {
    1: 'Simulation Modelling',
    2: 'Imaging/Imagery',
    3: 'Ecological/Landscape',
    4: 'Deep Learning',
    5: 'Material Mechanics',
    6: 'Fluid Mechanics',
    7: 'Signal Processing',
    8: 'Medical/Biology',
    9: 'Urban/Landscape',
    10: 'Misc/Structure'
}
```

```python
fig_cols = 2
fig_rows = round(topic_results.num_topics / fig_cols)

plt.tight_layout()
fig, axes = plt.subplots(fig_rows, fig_cols, figsize=(12,26))
plt.suptitle("Topic Trends", y=0.92, fontsize=22)
for t_id, ax_ in enumerate(axes.flatten()):
    topic_id = t_id + 1
    topic_ = topic_results.get_topic_by_id(topic_id)
    topic_.groupby('year').count()['title'].plot(kind='bar', 
                                                 color='#2b7bba', 
                                                 title='{}\nNum Publications: {}'.format(topic_names[topic_id], len(topic_.index)),
                                                 ax=ax_,
                                                 sharex=True,
                                                 sharey=True
                                                )
# End for
```

```python
fig.savefig(FIG_DIR+'initial_topic_trends.png', dpi=300)
```
